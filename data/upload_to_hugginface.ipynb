# Push the parquet file to huggingface Rodan009/pokemon-rag

```python
from datasets import Dataset
from pathlib import Path

# Config
dataset_path = Path("dataset/pokemon_data.parquet")
repo_id = "Rodan009/pokemon-rag"

# Check if file exists
if not dataset_path.exists():
    raise FileNotFoundError(f"{dataset_path} not found. Please run generate_parquet.py first.")

# Load Parquet
ds = Dataset.from_parquet(str(dataset_path))
print(f"Loaded {len(ds)} records from {dataset_path}")

# Push to Hub
print(f"Pushing to {repo_id}...")
try:
    ds.push_to_hub(repo_id, config_name="pokemon-knowledge", split="train")
    print("Successfully uploaded to Hugging Face Hub!")
except Exception as e:
    print(f"Error uploading to Hub: {e}")
    print("Ensure you are logged in with 'huggingface-cli login' or have set the HF_TOKEN environment variable.")
```
